# Logistic Regression 

```python
#creating a random dataset

from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
x,y = make_blobs(n_samples = 500,n_features = 2, centers =2, cluster_std = .7, random_state = 0)
```

```python
x.shape
```

```python
y.shape
```

```python
np.unique(y)
```

```python
#plotting the points
plt.scatter(x[:,0],x[:,1],c=y,cmap = "cool")
plt.show()
```

```python
#now performing the classification

from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=.20, random_state=42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(xtrain,ytrain)
```

```python
predict = log.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
test_score = accuracy_score(ytest,predict)
```

```python
train_score = accuracy_score(ytrain , log.predict(xtrain))
```

```python
print(" The Train Score Is :",train_score)
print()
print(" The Test Score Is :",test_score)
```
