```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# setting style
sns.set(style ='darkgrid')
```

```python
# loading the data

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Now we will see the relational plot 
sns.relplot(x='total_bill',y = 'tip',data =tips)
```

```python
# note by default the kind of plot is the scatter plot
```

```python
# facetgrid means it can plot so many axis on a single figure
```

```python
dir(sns.FacetGrid)
```

```python
# in the above graph we can see the linear relation between the totaL_bill and tip given
```

```python
# performing seperation on the basis of sex
sns.relplot(x ='total_bill',y = 'tip' , data =tips , hue ='sex')
```

```python
# performing seperation on the basis of smoker
sns.relplot(x ='total_bill',y ='tip' , data=tips , hue ='smoker')
```

```python
# we can see how smoker and non smoker have given the tips
```

```python
# determining the unique values present in the smoker category
tips['smoker'].value_counts()
```

```python
# determining the unique values present in the sex category
tips['sex'].value_counts()
```

```python
# now we want to provide more styling based on some categorical variable

sns.relplot(x='total_bill',y ='tip',data = tips , hue ='sex',style ='time')
```

```python
# but it's looking too complex
```

```python
sns.relplot(x='total_bill',y = 'tip',hue='size',data=tips)
```

```python
#for changing the size of the circle
sns.relplot(x='total_bill' , y ='tip', data =tips,size = 'size')
```

```python
# hue changes the color
# size changes the size of the object
#style changes the style of the represented object
```

```python
# if we want to provide the range to these sizes
sns.relplot(x='total_bill',y='tip',data =tips,size = 'size',sizes =(15,200))
```

```python
# so in the above we provided a range by using a sizes as an option
```

```python
# now let's see how we can work with the line plot 

# in the relplot there is one option called as kind and by default kind = 'scatter' 

# so we need to change the kind type
```

```python
from numpy.random import randn
```

```python
# creating a dataset

df = pd.DataFrame(dict(time = np.arange(500),value =randn(500).cumsum()))
```

```python
df.head()
```

```python
# now for the linear plot we will set the option kind = 'line'
```

```python
sns.relplot(x = 'time' , y ='value',kind ='line',data = df)
```

```python
# so now we have got a linear plot
```

```python
# in the x axis we had the time as a parameter which was already sorted well
```

```python
# now let's understand the role of sort()
```

```python
# creating a new dataset
```

```python
df = pd.DataFrame(randn(500,2).cumsum(axis =0),columns = ['time','value'])
```

```python
df.head()
```

```python
# now here nothing is sorted , everything is randomly present
```

```python
sns.relplot(x ='time', y ='value',data=df, kind ='line',sort =True)
```

```python
# this is the output when the x axis is sorted in nature
```

```python
sns.relplot(x ='time', y ='value',data=df, kind ='line',sort =False)
```

```python
# here the sorting is not performed
```

```python
# loading another dataset
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
# categorical variables present in the dataset 
# subject
# event
# region
```

```python
sns.relplot(x='timepoint',y='signal',kind = 'line',data =fmri)
```

```python
# so we can see two lines around the linear line this region shows the confidence interval
```

```python
# now for removing the confidence interval
```

```python
sns.relplot(x = 'timepoint',y='signal',kind ='line',data = fmri,ci = False)
```

```python
# ci == Confidence Interval
```

```python
# plotting the data with the semantic mapping
```

```python
sns.relplot(x='timepoint',y='signal',hue='event',kind = 'line',data =fmri)
```

```python
sns.relplot(x='timepoint',y='signal',hue ='region',style = 'event',kind ='line',data =fmri)
```

```python
# if we add hue and style then we get color change and style change
```

```python
# for plotting the multiple measurements together
sns.relplot(x='timepoint',y ='signal',hue ='region',units ='subject',estimator=None,kind='line',data = fmri.query("event=='stim'"))
```

```python
# loading another dataset
dots = sns.load_dataset('dots').query("align=='dots'")
```

```python
dots.head()
```

```python
sns.relplot(x='time',y='firing_rate',data=dots,kind='line')
```

```python
#let's create a color pallet
```

```python
palette = sns.cubehelix_palette(light =0.5,n_colors=6)
```

```python
sns.relplot(x='time',y='firing_rate',data =dots,kind='line',hue='coherence',style='choice',palette=palette)
```

```python
# we can also change the width of these lines based on intensity
```

```python
sns.relplot(x='time',y='firing_rate',size ='coherence',style='choice',kind='line',data=dots)
```

```python
# since we used size='coherence' that's why based on the value the size of the line changes
```
